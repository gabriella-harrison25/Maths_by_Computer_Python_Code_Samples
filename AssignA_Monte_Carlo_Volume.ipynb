# Monte Carlo Volume Approximations
*Purpose:* Perform a Monte-Carlo approximation to the volume of a cylinder with radius one and height one

*Dependencies:* numpy, math

*Functions:* 

Indicator_cylinder(x,y,z)
- Input: (x,y,z) are coordinates of a point
- Return: 1 if point in cylinder; 0 if point not in cylinder

Estimate(N)
- Input: N is positive integer of number of points
- Return: Volume of cylinder

*To Run:* python AssignA_Monte_Carlo_Volume.py

**Theoretical Approach**

For a cylinder of radius 1 and height 1, we know that $$V = \pi \times r^2 \times h$$ Therefore, the volume we should find is: $$V = \pi (1)^2(1) = \pi$$

```python
#imports for notebook
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#the indicator function
def Indicator_cylinder(x,y,z):
    h = 1 #height of cylinder
    R = 1 #radius of cylinder base
    r = np.sqrt(x**2+y**2) #radius of base of cylinder
    return 1*(r<R and z<h)
```

```python
#the estimate function (system modeled after aforementioned notebook)
def Estimate(N):
    #Monte-Carlo Method

    #generate random points in [x_min, x_max] x [y_min, y_max] x [z_min,z_max]
    x_min, x_max = -1,1
    y_min, y_max = -1,1
    z_min, z_max = 0,1

    x = np.random.uniform(x_min, x_max, N)
    y = np.random.uniform(y_min, y_max, N)
    z = np.random.uniform(z_min, z_max, N)

    #add up their values
    total = 0
    for i in range(N):
        total += Indicator_cylinder(x[i],y[i],z[i])

    #determine volume
    cuboid_volume = (x_max-x_min)*(y_max-y_min)*(z_max-z_min)
    cylinder_volume = cuboid_volume *total/N

    print(f"Volume of object for {N} points is", cylinder_volume)
```

```python
for N in [100, 1000, 10000]:
    Estimate(N)
```
