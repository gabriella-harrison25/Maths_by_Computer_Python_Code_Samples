# Taylor Series Approximations
*Purpose:* Calculates and visualizes numerical polynomial approximations to sine function. Graph displays $x \in [-3\pi, 3 \pi]$ for approximations with between one and seven terms.

*Dependencies:* numpy, matplotlib, math

*To Run:* python AssignA_Taylor_Series_Approx.py

```python
#imports for notebook
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#defining the function that evaluates the approximation at a given point
def taylorsin(x,N):
    terms = []
    for n in range(1, N+1):
        term = (-1)**(n+1)*x**(2*n-1)/math.factorial(2*n-1)
        terms.append(term)
    sum_of_terms = sum(terms)
    return sum_of_terms
```

```python
#testing it works appropriately (within some error margin)
taylorsin(0.5, 2)
```

```python
#define your x-values
x_values = np.linspace(-3*np.pi, 3*np.pi, 200)

#plot the original sin(x)
plt.plot(x_values, np.sin(x_values), label = "sin(x) (Actual)", color = "black", linestyle = "--")

#iterate over number of terms so the approximations get plotted
for N in range(1,8):
    y_approx = [taylorsin(x_val, N) for x_val in x_values]
    plt.plot(x_values, y_approx, label = f"Taylor ({N} terms)")

#specific characteristics of the plot
plt.title("Taylor Approximations of sin(x)")
plt.xlabel("x(radians)")
plt.ylabel("y")
plt.legend(bbox_to_anchor=(1.05, 0.5), loc="center left")
plt.grid(True)
plt.xlim(-3*np.pi, 3*np.pi)
plt.ylim(-3, 3)
plt.show()
```
