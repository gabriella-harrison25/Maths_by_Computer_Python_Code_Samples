```python
#Import packages
import networkx as nx 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#loading the zebra data to make a graph
fh = open("out.zebra", "rb")
zebra_graph = nx.read_edgelist(fh, comments = '%')
fh.close()
print(type(zebra_graph))
```

```python
#finding number of edges and nodes in the graph
print("number of nodes:", zebra_graph.number_of_nodes())
print("number of edges:", zebra_graph.number_of_edges())
```

```python
#generate layout for nodes
positions = nx.spring_layout(zebra_graph)

#visualize network
nx.draw(zebra_graph, positions)
```

## Observation about graph
There are two major clusters of nodes. One of them consists of four nodes, and the other consists of the other 23 nodes. There is a substantial difference in the size of the clusters, indicating there must be an attribute causing such separation.

```python
#adjacency matrix practice (not required, but i was curious)
A = nx.to_numpy_array(zebra_graph)
plt.spy(A)
```

# Observation
There is a high likelihood of edges between values in the top left quadrant of the data. Similar to the visualization above, there are two primary clusters present. One of them is substantially larger than the other.
