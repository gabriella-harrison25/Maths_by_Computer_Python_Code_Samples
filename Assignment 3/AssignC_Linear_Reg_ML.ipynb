# Linear Regression Machine Learning Model
*Purpose:* Create a linear regression machine learning model to generate a linear regression based on the given dataset.

*Dependencies:* numpy, matplotlib, sklearn

*To Run:* python AssignC_Linear_Reg_ML.ipynb

*Context on Dataset:*

An array $X$ and an array $y$ are created. $X$ will be the design matrix and $y$ the target in a machine learning task. $X$ is a $100\times 2$ matrix and $y$ is a $1\times 100$ matrix.  Row $i$ of $X$ contains real numbers $x_{i,1},x_{i,2}$ and these have been used to create the value $y_i$ in the corresponding position of $y$ via the formula 

$$y_i = 3x_{i,1}+2x_{i,2}+4 + \mu$$ 

where $\mu$ is sampled, independently for each $i$, from a normal distribution with mean of $0$ and variance of $1$. 

The given code generates a plot of the points $(x_{i,1},x_{i,2},y_i)$ in three dimensional space. Most are blue but the points corresponding to the last 5 rows of $X$ and the last five entries in $y$ are shown in red.

```python
#MY CODE/SOLUTION (I split up the given code into different boxes)
#import packages
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import linear_model
```

```python
#given code
n = 100

x1 = np.random.uniform(0,4,n)
x2 = np.random.uniform(0,4,n)
y = 3*x1 + 2*x2 + 4 + np.random.normal(0,1,n)

X = np.array([x1,x2]).transpose() #creates the X matrix as an n by 2 matrix with each row a point (x1,x2)
```

```python
#my code
#splitting matrix into training/testing (last five rows for testing)
X_train= X[:-5, :]
X_test = X[-5:, :]

y_train = y[:-5,]
y_test = y[-5:]

#select and train model
regr = linear_model.LinearRegression()
regr.fit(X_train,y_train)
```

```python
#printing linear regression coefficients
a = regr.coef_[0]
b = regr.coef_[1]
c = regr.intercept_
print("For the function f(x_1, x_2), the coefficients a, b, c are:",a,b,c)
```

```python
#my code
#predictions
y_predicted = regr.predict(X_test)

#print the coefficients we find based on the prediction
print("Predicted values:", y_predicted)
print("Test values:", y_test)
```

```python
#given code
# initialise 3D plotting
fig = plt.figure(figsize=(8,8))
ax = plt.subplot(projection='3d')

#MY CODE
X = np.linspace(0,4, 100)
Y = np.linspace(0,4,100)

X,Y = np.meshgrid(X,Y)
Z = a*X+b*Y+c

ax.plot_surface(X,Y,Z, color = 'green', alpha = 0.5)

#given code
ax.plot(x1[:-5],x2[:-5],y[:-5], "b.", lw = 3)
ax.plot(x1[-5:],x2[-5:],y[-5:], "r.", lw = 3)

ax.set_xlabel("x", fontsize=14)
ax.set_ylabel("y", fontsize=14)
ax.set_zlabel("z", fontsize=14)
plt.show()
```
