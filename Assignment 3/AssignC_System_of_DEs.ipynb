# Systems of Differential Equations - Modelling
*Purpose:* Solve the following system of differential equations using scipy. Plot the solutions' phase portrait.
$$
\begin{align}
\dot y_1 & = -y_2  \\
\dot y_2 & = y_1
\end{align}
$$
$y_1,y_2$ are functions of t for $0\leq t \leq \pi$ with initial conditions $y_1(0)=3, y_2(0)=0$

*Dependencies:* numpy, matplotlib, scipy, odeint

*Functions:*
system(y,t):
- Input: y,t, floats
- Output: Defines system of differential equations to solve


*To Run:* python AssignC_System_of_DEs.ipynb

```python
#import relevant packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
#write system of equations as a function
def system(y,t):
    y1, y2 = y
    y1dot = -y2
    y2dot = y1
    return y1dot, y2dot
```

```python
#set up the problem
#initial conditions
y0= np.array([3,0])

#time grid
Nsteps = 1000
tf=np.pi
t = np.linspace(0,tf, Nsteps+1)
```

```python
y_solution = odeint(system, y0, t)

y1 = y_solution[:, 0]
y2 = y_solution[:,1]
```

```python
#plotting phase portrait
plt.figure(figsize = (5,5))
plt.plot(y1,y2)
plt.plot(y1[0], y2[0], 'd', markersize = 10, label = "t = 0") 
plt.plot(y1[-1], y2[-1], 'o', markersize = 10, label = "t= $\pi$")

plt.xlim(-6,6)
plt.ylim(-6,6)
plt.xlabel("$y_1$", fontsize = 14)
plt.ylabel("$y_2$", fontsize = 14)
plt.title("Phase Portrait for System of ODEs", fontsize = 16)
plt.legend()
plt.show()
```
