```python
#Run this cell once you have put wineequality-white.csv
#into the same folder as you have this notebook

import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd

wine = pd.read_csv("winequality-white.csv", sep=';') 
wine.head()
```

```python
#create series for fixed acidity

Series = wine["fixed acidity"]
Series.head()
```

```python
#plotting histogram
plt.hist(Series, bins = 40)
plt.xlabel("Fixed Acidity")
plt.ylabel("Frequency")
plt.title("Histogram of Fixed Acidity Data")
plt.show
```

$\textbf{Observation about the Histogram: }$
The majority of the data lies between an acidity of 6-8. It appears somewhat like a normal distribution centered around a fixed acidity of 7.
