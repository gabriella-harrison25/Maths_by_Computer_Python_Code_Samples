# Monte Carlo Integral Approximations
*Purpose:* Perform a Monte-Carlo approximation to a definite integral of $f(x)=\sqrt{1-x^4}$ on $[0,1]$

*Dependencies:* numpy, matplotlib, math

*To Run:* python AssignA_Monte_Carlo_Integrals.py

```python
#imports for notebook
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#modeled after the code in previous assignments

#define the function to integrate
def function(x):
    return np.sqrt(1-x**4)

#define the interval and the number of samples(points)
a,b = 0,1
N = 900

#generate samples and evaluate f at each
x = np.random.uniform(a,b,N)
f_of_x = function(x)

#compute mean of function
f_mean = np.mean(f_of_x)

#compute integral estimate based on mean value
integral = (b-a)*f_mean
print(f"The Monte-Carlo estimate for the integral using {N} samples is {integral}")

x_plot = np.linspace(a, b, 101)
y_plot =function(x_plot)
plt.plot(x_plot ,y_plot, 'b', label = "Function") 
plt.plot(x, f_of_x, 'ro', label = "Approximated Function")
plt.plot([a,b,b,a,a],[0,0,f_mean, f_mean,0], 'g', label = "Approximating Area Rectangle")
plt.legend(bbox_to_anchor = (0.25, 0.2), loc = "center left")
plt.show()
```
