# Series Convergence - Analysis

*Purpose:* Determine the number of terms required for the series $\sum_{i=0}^\infty (-1)^{n+1} \frac{1}{n}$ to get within a particular value $e$ of $\ln(2)$ (which it converges to).

*Dependencies:* numpy

*Functions*

howfar(e)
- Input (e): positive float
- Output: N, positive integer, the minimum number of terms such that series is within e of convergence value

*To Run:* python AssignB_Series_Convergence.ipynb

```python
import numpy as np
```

```python
#setting up partial sums function
log2 = np.log(2)

def partial_sums(terms):
    n = np.arange(terms+1)
    Sn = np.zeros(terms+1)
    for i in range(1, terms+1):
        Sn[i]=Sn[i-1] + (-1)**(i+1)*(1/i)
    return Sn[terms]
```

```python
#creating the howfar function
def howfar(e):
    n = 1
    while abs(log2-partial_sums(n))>=e:
        n = n +1
    print("Smallest N such that the distance to ln(2) is less than epsilon =", e, "is", n)
```

```python
#using howfar for the specific values of epsilon asked for
howfar(0.1)
howfar(0.01)
howfar(0.001)
```
