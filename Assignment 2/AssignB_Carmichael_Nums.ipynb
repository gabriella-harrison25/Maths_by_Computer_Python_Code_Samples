# Carmichael Numbers - Foundations

*Purpose:* Determine if a particular number is a Carmichael number, and list all the Carmichael numbers between given integers.

*Dependencies:* numpy

*Functions*

euclideanhcf(m,n):
- Input: m,n positive integers
- Output: highest common factor of the two inputs

isprime(n):
- Input: n, a positive integer
- Output: boolean, True if input is a prime number

carmichael(n):
- Input: n, a positive integer
- Output: boolean, True if n is a carmichael number


*To Run:* python AssignB_Carmichael_Nums.ipynb

```python
import numpy as np
```

```python
#define gcd = hcf function for the def of Carmichael
def euclideanhcf(m,n):
    a = min(m,n)
    b=max(m,n)
    r = b%a
    if r==0:
        return a
    q=b//a
    while r !=0:
        prevr = r
        b=a
        a=r
        r=b%a
        q=b//a
    return prevr
```

```python
#define prime testing function to avoid these in list of carmichaels
def isprime(n):
    prime = True
    for i in range(2, n):
        if n%i ==0:
            prime = False
    return prime
```

```python
#finding carmichael numbers function
def carmichael(n):
    if isprime(n) == True:        #test if input is prime
        return False      #if it is prime, it cannot be carmichael
    if isprime(n)==False:         #if input not prime, then we need to do carmichael test
        for a in range(1,n):       #test for all values of a
            if euclideanhcf(a,n)==1:
                u = n-1
                v = a**u
                w = v%n              #testing equation above (a^(n-1) = 1 mod n)
                if w != 1:
                    return False
        return True
```

```python
#show the function above returns true if carmichael number is true
#and false otherwise
for n in range(561, 563):
    print("The statement '", n, "is Carmichael' is", carmichael(n))
```

```python
#complete list of carmichael numbers from 2 to 2821 inclusive
carmichael_numbers = []
for i in range(2,2822):
    if carmichael(i) ==True:
        carmichael_numbers.append(i)

print(carmichael_numbers)
```
