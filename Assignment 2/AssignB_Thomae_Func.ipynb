# Thomae Function - Analysis
*Purpose:* Prove that the Thomae Function is only continuous at points $c \in (0,1)$ where $c$ is irrational utilizing numerical approximations.

*Dependencies:* numpy

*Functions:*

higher(e)
- Input: e is positive float $0<e<1$.
- Output: number of values $x \in (0,1)$ where $f(x)\geq e$.

*To Run:* python AssignB_Thomae_Func.ipynb

## Background
The **Thomae Function** is given by: 
$f:(0,1) \rightarrow \mathbb{R}$ be defined by $$f(x)=\left\{\begin{array}{lr}
       \displaystyle\frac{1}{q} & \text{if } x\in\mathbb{Q}\ x = \displaystyle\frac{p}{q}\ p,q\in\mathbb{Z}, q>0, \text{ hcf}(p,q)=1.\ \\
        0 & \text{if } x\not\in\mathbb{Q}
        \end{array}\right. \text{.}$$

It is continuous at $c$ whenever $0<c<1$ is irrational. The code below is written to show there are only finitely many values of $x$ in $[0,1]$ where the function is discontinuous. 

```python
#notebook imports
import numpy as np
```

```python
#defining the Thomae function on (0,1):
x_list = []
y_list = []

for q in range(1,500):
    for p in range(1,q):
        if np.gcd(p,q)==1:
            x_list += [p/q] 
            y_list += [1/q]
```

```python
#defining the higher function
def higher(e):
    if 0<e<1:
        limit = e
        number_of_values = sum(y>= limit for y in y_list)
        return number_of_values
    else:
        print("e not in range 0<e<1")
```

```python
#finding number of values for various e's
for e in [0.5,0.25,0.125]:
    print("For e =", e, "there are", higher(e), "values where f(x)>=e")
```
