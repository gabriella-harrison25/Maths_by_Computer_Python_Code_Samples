# Cyclic Groups - Algebra

*Purpose:* Determine, given a positive integer n, if the unit group of integers modulo n is cyclic.

*Dependencies:* numpy

*Functions*

euclideanhcf(m,n):
- Input: m,n positive integers
- Output: the highest common factor of m and n

z_modn(n):
- Input: n, a positive integer
- Output: generates modulo group

order(n,m):
- Input: n,m, positive integers
- Output: the order of an element m in set Z/nZ

iszncyclic(n):
- Input: n, positive integer
- Output: boolean, True if unit group is cyclic

```python
import numpy as np
```

```python
#define gcd = hcf function for the def of modulo groups
def euclideanhcf(m,n):
    a = min(m,n)
    b=max(m,n)
    r = b%a
    if r==0:
        return a
    q=b//a
    while r !=0:
        prevr = r
        b=a
        a=r
        r=b%a
        q=b//a
    return prevr
```

```python
#define function that creates the modulo group
def z_modn(n):
    z_modn = []
    for a in range(1, n):
        if euclideanhcf(a,n)==1:
            z_modn.append(a)
    return z_modn
```

```python
#define order function (finds order of an element)
def order(n,m):
    if m>=n:
        print("Error: m>n")
    if euclideanhcf(n,m)!=1:
        print("Not an element of set")
    else:
        order = 1
        remainder_now = m%n
        while remainder_now !=1:
            remainder_now = (remainder_now*m)%n
            order= order+1
        return order
```

```python
#define function testing cyclicality
def iszncyclic(n):
    is_cyclic = False
    group = z_modn(n)
    group_order = len(group)  #also finds order of group

    for g in group:
        if order(n,g) == group_order:   #group is cyclic if an element in that group has order of group itself
            is_cyclic = True
    return is_cyclic
```

```python
#print list of integers n from 2 to 50 inclusive where z/nz is cyclic
for n in range(2,51):
    if iszncyclic(n) ==True:
        print(n)
```
